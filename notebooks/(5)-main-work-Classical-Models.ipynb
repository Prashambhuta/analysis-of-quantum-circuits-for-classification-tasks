# Classical ML Running, Multiclass Single Label Prediction

```python
# Imports
import numpy as np
import pandas as pd
import time
```

```python
# generate fake data
from sklearn.datasets import make_classification, make_multilabel_classification
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import accuracy_score, f1_score
```

```python
# for data split
from sklearn.model_selection import train_test_split
from qiskit_algorithms.utils import algorithm_globals
algorithm_globals.random_seed = 123
from pycaret.classification import *
```

```python
class classicalAlgoTraining:
    """
    Class to train algorithms."""
    def __init__(self, no_of_features, no_of_samples, no_of_classes, models = ['knn', 'qda', 'rf', 'lightgbm', 'svm']):
        self.no_of_features = no_of_features
        self.no_of_samples = no_of_samples
        # self.no_of_labels = no_of_labels
        self.no_of_classes = no_of_classes
        self.models = models
        print(f"the no of classes is {self.no_of_classes}")

    def data_generation(self):
        """
        Generate classification data using sklearn's data generation.\
        """
        n_samples=self.no_of_samples
        n_features=self.no_of_features
        n_classes=self.no_of_classes
        print(n_samples)
        X, y = make_classification(n_samples=n_samples,
                                   n_features=n_features,
                                   n_classes=n_classes,
                                    n_redundant=0,
                                    n_clusters_per_class=1,
                                    class_sep=2.0,
                                    random_state=algorithm_globals.random_seed
                                    )
        print(f"the number of classses is: {n_classes}")
        return X, y

    # def train_test_split(self):
    #     """
    #     splitting the data into train test split
    #     """
    #     ## Train test split
    #     # X, y = self.data_generation()
    #     train_features, test_features, train_labels, test_labels = train_test_split(
    #         X, y, train_size=0.8, random_state=algorithm_globals.random_seed
    #     )
    #     return train_features, test_features, train_labels, test_labels

    def pycaret_training(self):
        X, y = self.data_generation()
        print(X.shape, y.shape)
        exp_name = setup(data = X,  target = y, train_size=0.8)
        best_model = compare_models(include=self.models, sort='f1')
        # print(best_model)
        results = {}
        for model in self.models:
            model_created = create_model(model, verbose=False)
            model_results = predict_model(model_created)
            short_results = pull(model_created)
            # print(pull(model_created).columns)
            results.setdefault('Model', []).append(short_results['Model'].values[0])
            results.setdefault('Accuracy', []).append(short_results['Accuracy'].values[0])
            results.setdefault('F1', []).append(short_results['F1'].values[0])
            # print(results)
            # pull(model_created).to_csv(f'''results/{model}-features_{self.no_of_features}-samples_{self.no_of_samples}-classes_{self.no_of_classes}.csv''', index=False)
        df = pd.DataFrame(results)
        df.to_csv(f'''results/MC-SL-features_{self.no_of_features}-samples_{self.no_of_samples}-classes_{self.no_of_classes}.csv''', index=False)
        return df
```

```python
no_of_features = [2, 4, 6, 8, 10]
no_of_samples = [64, 128, 256, 512, 1024, 2048]
no_of_classes = [3, 4, 5]
```

```python
# 2-feature, 10-samples, 2- class training
for clas in no_of_classes:
    for feature in no_of_features:
        for sample in no_of_samples:
            training_object = classicalAlgoTraining(feature, sample, clas)
            training_object.pycaret_training()
```

```python
results = training_object.pycaret_training()
```
